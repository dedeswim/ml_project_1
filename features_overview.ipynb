```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Import data
filename = 'data/train.csv'
raw_df = pd.read_csv(filename)

size = raw_df.shape[0]
print('Total size: ' + str(size))

raw_df.head()
```

```python
def compute_missing_values(df, column_name):
    
    column = df[column_name].to_numpy()

    return column[np.where(column == -999.0)].shape[0]
```

```python
for column in raw_df.columns[2:]:
    n_nan = compute_missing_values(raw_df, column)
    rel_nan = n_nan / size
    
    print('{column} is missing {n_nan} values, with a percentage of {rel_nan}'
              .format(column=column, n_nan=n_nan, rel_nan=rel_nan)
         )
```

We can see that several features have high percentages of missing values (> 40%), then we start by training the data without those features. Based on the information we will understand, we will next decid how to treat the `DER_mass_MMC` feature, which is missing ~15% of the values.

```python
def nan_columns(df):
    
    columns = []
      
    for column in df.columns[2:]:
        n_nan = compute_missing_values(df, column)
        rel_nan = n_nan / size

        if rel_nan > 0.3:
            columns.append(column)
    
    return columns
```
